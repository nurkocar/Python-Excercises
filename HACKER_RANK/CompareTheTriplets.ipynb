```python
alice = [12, 20, 3]
bob = [10, 54, 8]
a = 0
b = 0
```

```python
for i in range(len(alice)):
    if alice[i] > bob[i]:
        a += 1
    elif bob[i] > alice[i]:
        b += 1
print(a, b)
```

```python
def compareTriplets(a, b):
    score_a = score_b = 0
    for i in range(len(a)):
        if a[i] > b[i] : score_a += 1
        elif b[i] > a[i] : score_b += 1
    return score_a, score_b
```

```python
compareTriplets([5, 6, 7], [3, 6, 10])
```
